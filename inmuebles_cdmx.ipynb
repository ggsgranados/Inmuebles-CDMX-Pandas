## Remover registros inconsistentes

```python
# Consultando los registros donde el valor o el condominio sean cero
df.query('Valor == 0 | Condominio == 0')
```

```python
# Tomando los indices de los registros
df.query('Valor == 0 | Condominio == 0').index
```

```python
# Creando el filtro
df_remover = df.query('Valor == 0 | Condominio == 0').index
```

```python
# Eliminando los registros
# Argumentos (indices a remover, eje, inplace=True para que quede guardado)
df.drop(df_remover, axis=0, inplace=True)
```

```python
# Visualizando parte del dataframe
df.head()
```

```python
# Se observa que, como todos son del Tipo Departamento, la columna es innecesaria, se procede a borrar
df.Tipo.unique()
# Eliminando la columna Tipo
df.drop('Tipo', axis=1, inplace=True)
```

## Creando filtros

### 1. Apartamentos que tienen 1 dormitorio y un alquiler menor a MXN 4200.

```python
# Creando el filtro
filtro1 = (df['Habitaciones'] == 1) & (df['Valor'] < 4200)
# Aplicando el filtro al dataframe y creando otro dataframe
df1 = df[filtro1]
df1.head()
```

```python
# Comprobando que el fintro ha sido aplicado
df[filtro1].max()
```

### 2. Apartamentos que tienen al menos 2 dormitorios, un alquiler menor a MXN 10500 y una superficie mayor a 70 m².

```python
# Creando el filtro
filtro2 = (df['Habitaciones'] >= 2) & (df['Valor'] < 10500) & (df['Area'] > 70)
# Aplicando el filtro al dataframe y creando otro dataframe
df2 = df[filtro2]
df2.head()
```

## Guardando los archivos

```python
# Argumentos (nombre que tendra el archivo, elimina el indice anterior, separador)
df.to_csv('inmuebles_residenciales.csv', index=False, sep=';')
```

# Demandas de DEV

## Crear columnas numéricas

### 1. valor_mensual: Esta columna debe contener los gastos mensuales de cada propiedad, incluyendo el alquiler y el condominio.

```python
# Datos originales
datos = pd.read_csv(url, sep= ';')
datos.shape
```

```python
# Creando la nueva columna
datos['valor_mensual'] = datos['Valor'] + datos['Condominio']
datos.head()
```

### 2. valor_anual: Esta columna debe contener los gastos anuales por propiedad, es decir, el Impuesto más 12 meses de alquiler y condominio.

```python
datos['valor_anual'] = datos['valor_mensual']*12 + datos['Impuesto']
datos.tail()
```

## Crear columnas categoricas

### Descripcion: Esta columna debe contener un resumen de la información clave de las propiedades que se mostrarán en el sitio web: tipo de propiedad, barrio, cantidad de habitaciones y plazas de estacionamiento.



```python
datos['Descripcion'] = datos['Tipo'] + ' en ' + datos['Colonia'] + ' con ' + datos['Habitaciones'].astype(str) + ' habitaciones y ' + datos['Garages'].astype(str) + ' plazas de estacionamiento.'
datos.head()
```

### Tiene_suite: Esta debe ser una columna que indique únicamente si la propiedad tiene o no suites, sin importar la cantidad.

```python
datos['Tiene_suite'] = datos['Suites'].apply(lambda x: 'Tiene' if x > 0 else 'No tiene')
datos.head()
```

```python
# Guardando el dataframe
datos.to_csv('inmuebles_dev.csv', index=False, sep=';')
```
